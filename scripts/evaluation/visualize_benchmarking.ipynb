```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import matplotlib as mpl
mpl.rcParams['font.family'] = 'Times New Roman'

# Full dataset
data = {
    "Model": ["GCN", "GAT", "TransConv", "PNC", "FC", "XGBoost"],
    "Params (M)": [1.31, 1.32, 1.32, 1.65, 5.35, 5.35],
    "MSE": [26.22, 21.79, 20.57, 20.71, 72.77, 72.77],
    "Inference Time (ms)": [19.05, 22.45, 17.18, 30.45, 0.46, 0.46]
}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x="Params (M)",
    y="MSE",
    hue="Model",
    palette="tab10",
    s=150  # Fixed dot size
)

# Add labels to each point
for i in range(df.shape[0]):
    plt.text(df["Params (M)"][i] + 0.05, df["MSE"][i], df["Model"][i], fontsize=9)

# plt.title("Model Complexity vs Mean Squared Error", fontsize=14, fontfamily="Times New Roman")
plt.xlabel("Number of Parameters (Millions)", fontsize=14, fontfamily="Times New Roman")
plt.ylabel("Mean Squared Error (MSE)", fontsize=14, fontfamily="Times New Roman")
plt.xticks(fontsize=14, fontfamily="Times New Roman")
plt.yticks(fontsize=14, fontfamily="Times New Roman")
plt.xlim(left=1)
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, title="Model", fontsize=14, title_fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Sort by inference time
df_sorted = df.sort_values("Inference Time (ms)", ascending=True)

plt.figure(figsize=(8, 4))
sns.barplot(x="Inference Time (ms)", y="Model", data=df_sorted, palette="crest")
plt.title("Inference Time by Model", fontsize=13, fontfamily="Times New Roman")
plt.xlabel("Inference Time (ms)")
plt.ylabel("Model")
plt.tight_layout()
plt.show()
```
